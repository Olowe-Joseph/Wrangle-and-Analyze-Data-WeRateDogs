The goal of data wrangling is to assure quality and useful data. The default data provided, and the data that was gathered querying twitter Api is dirty and messy. the wrangling process which is basically gathering, accessing and cleaning is what has helped in making our data look good and presentable. Cleaning is a crucial part of data wrangling which helps transform the data from a dity and messy data to what we have not which is easily understandable to a layman. A total of 8 quality issues and 2 tidyness issues were cleaned in this project. that's not all though, as there are some issues which were not mentioned in this project but if we wrangle further, we would find them. As we know that there are three requirements for tidyness and this has to be strictly followe. they are:
1.Each variable forms a column
2.Each observation forms a row
3.Each type of observational unit forms a table.

This rule is what should be kept in mind when handling tidyness issues. the approach is always different, here we are stating the issue, defining how to handle it and we are writing our cleaning code.
